```python
import pandas as pd
import os
```

```python
%%time
NEW_DATA_PATH = '../ukbb_image_data.txt'

BODY_SURFACE_PATH = '../ukbb_body_surface.txt'
df = pd.read_csv(NEW_DATA_PATH, sep='\t')

body_surface_data = pd.read_csv(BODY_SURFACE_PATH, sep='\t')
df = df.merge(body_surface_data,on='eid',how='left')

new_df = df.copy()
cols_second_visit = new_df.columns[[new_df.columns.str.endswith('2.0')]].to_list()
cols_second_visit = ['eid'] + cols_second_visit

# Initialize an empty dictionary
col_name_dict = {}

# Open the file and read line by line
with open('../image_columns_info.txt', 'r') as file:
    for line in file:
        # Split the line by the tab character
        info = line.strip().split('\t')
        key = info[0]
        value = info[1]
        
        # Add to the dictionary
        col_name_dict[key] = value
def remove_visit_info(name):
    if name!='eid':
        name = name.split('-')[0]
    return name

new_df = new_df[cols_second_visit]

new_df.columns = new_df.columns.map(remove_visit_info)

new_df.columns = ['eid']+ [col_name_dict[col] for col in new_df.columns if col!='eid']

target_columns = new_df.columns.to_list()
target_columns.remove('IDP pipeline version')

new_df = new_df[target_columns]
new_df = new_df.set_index('eid')
new_df
```
