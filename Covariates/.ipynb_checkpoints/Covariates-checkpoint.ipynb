```python
import pandas as pd
```

```python
array_df = pd.read_csv("../../../ukbb_arraytype.txt", sep='\t')
array_df.rename(columns={"22000-0.0":"Genotype measurement batch"},inplace=True)
imputed_pheno_df = pd.read_csv("../../data/imputed_pheno_df_shuan.csv")
arr_imputed_pheno_df = imputed_pheno_df.merge(array_df,on='eid',how="left")
arr_imputed_pheno_df
```

```python
covariates = ["eid","instance_2.age_when_attended_assessment_centre_f21003","instance_0.sex_f31","instance_0.standing_height_f50","instance_2.Mean_Weight","instance_2.Merged_Medication_for_cholesterol_blood_pressure_diabetes_or_take_exogenous_hormones.Blood_pressure_medication","Genotype measurement batch","instance_2.uk_biobank_assessment_centre_f54"]
covariates_df = arr_imputed_pheno_df[covariates]
covariates_df["FID"] = covariates_df["eid"]
covariates_df.rename(columns = {"eid":"IID"},inplace=True)
new_order = ["FID"] + [i for i in covariates_df.columns if i!="FID"]
# new_order
covariates_df = covariates_df[new_order]
```

```python
patient_ids = pd.read_csv("../../data/patient.imaged_filtered.txt",sep="\t",header=None).iloc[:,0].to_list()
```

```python
covariates_df = covariates_df[covariates_df["IID"].isin(patient_ids)]
covariates_df.to_csv("../../data/covariates_df.2024.06.14.csv",index=False)
```
